```python
import json

def split_json(input_file, output_prefix, chunk_size):
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Tính toán số lượng phần cần tạo
    num_chunks = (len(data) + chunk_size - 1) // chunk_size

    for i in range(num_chunks):
        start_idx = i * chunk_size
        end_idx = (i + 1) * chunk_size
        chunk_data = data[start_idx:end_idx]

        output_file = f"{output_prefix}_{i+1}.json"
        with open(output_file, 'w', encoding='utf-8') as f_out:
            json.dump(chunk_data, f_out, indent=2, ensure_ascii=False)

        print(f"Created {output_file} with {len(chunk_data)} elements")

# Sử dụng hàm để chia file json thành các file con
split_json('href_khcn.json', 'href_khcn', 30)
```
