```python
import json
import pandas as pd

with open('info.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

df = pd.DataFrame(data)

df['Ngày, tháng, năm ban hành văn bản'] = pd.to_datetime(df['Ngày, tháng, năm ban hành văn bản'], errors='coerce', format='%d/%m/%Y')

# Chuyển đổi cột 'Thời điểm có hiệu lực'
df['Thời điểm có hiệu lực'] = pd.to_datetime(df['Thời điểm có hiệu lực'], errors='coerce', format='%d/%m/%Y')

# In kết quả
df
```

```python
df.to_json('output.json', orient='records', lines=True, force_ascii=False)
```

```python
import pandas as pd

# Đọc file JSON chứa trường "Số, ký hiệu văn bản"
info_path = 'info/info.json'
info_data = pd.read_json(info_path, orient='records', encoding='utf-8')

# Lấy ra list các giá trị của trường "Số, ký hiệu văn bản"
list_1 = info_data['Số, ký hiệu văn bản'].tolist()

# Đọc file JSON chứa trường "id"
href_path = 'href/href.json'
href_data = pd.read_json(href_path, orient='records', encoding='utf-8')

# Lấy ra list các giá trị của trường "id"
list_2 = href_data['id'].tolist()

# Tìm các giá trị xuất hiện trong list 1 mà không xuất hiện trong list 2
values_not_in_list_2 = [value for value in list_1 if value not in list_2]

# In kết quả
print("Các giá trị xuất hiện trong list 1 mà không xuất hiện trong list 2:")
print(values_not_in_list_2)
```
